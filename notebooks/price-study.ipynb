```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# price per 1K tokens
davinci_1k_price = 0.02
turbo_1k_price = 0.002
ft_davinci_1k_price = 0.12
```

```python
# one question deals on average 800-1000 tokens (input/output combined)
# price per request = (totalTokens/1kToken) * model_price
price_per_request_davinci = (( 1800 / 1000 ) * davinci_1k_price)
price_per_request_turbo = (( 1800 / 1000 ) * turbo_1k_price)
ft_price_per_request_davinci = (( 800 / 1000 ) * ft_davinci_1k_price)
print(f'Davinci per request: {price_per_request_davinci} $')
print(f'Turbo per request: {price_per_request_turbo} $')
print(f'FineTuned per request: {ft_price_per_request_davinci} $')
```

```python
# let' say we have 20K customers a month ( 1/10 of OPorto Population )
# each deals 15 requests per month
monthly_requests = 300000
```

```python
# price of monthly requests
davinci_monthly = int(price_per_request_davinci * monthly_requests)
turbo_monthly = int(price_per_request_turbo * monthly_requests)
ft_monthly = int(ft_price_per_request_davinci * monthly_requests)
print(f'Davinci Monthly Expenses: ~ {davinci_monthly}')
print(f'Turbo Monthly Expenses: ~ {turbo_monthly}')
print(f'Fine Tuned Monthly Expenses: ~ {ft_monthly}')
```

```python
# price of 6 months worth of requests
davinci_half_year = davinci_monthly * 6
turbo_half_year = turbo_monthly * 6
ft_half_year = ft_monthly * 6
print(f'Davinci 6 months Expenses: ~ {davinci_half_year}')
print(f'Turbo 6 months Expenses: ~ {turbo_half_year}')
print(f'Fine Tuned 6 months Expenses: ~ {ft_half_year}')
```

```python
# price of annually requests
davinci_annually = davinci_monthly * 12
turbo_annually = turbo_monthly * 12
ft_annually = ft_monthly * 12
print(f'Davinci Annually Expenses: ~ {davinci_annually}$')
print(f'Turbo Annually Expenses: ~ {turbo_annually}$')
print(f'Fine Tuned Annually Expenses: ~ {ft_annually}$')
```

```python
davinci_data = [davinci_monthly, davinci_half_year, davinci_annually]
turbo_data = [turbo_monthly, turbo_half_year, turbo_annually]
ft_data = [ft_monthly, ft_half_year, ft_annually]

width = 0.3
labels = [1, 6, 12]
plt.xticks(range(len(davinci_data)), labels)

plt.bar(range(len(turbo_data)), turbo_data)
plt.bar(range(len(davinci_data)), davinci_data, bottom=turbo_data)
plt.bar(range(len(ft_data)), ft_data, bottom=davinci_data)

plt.title("Few shot vs Fine Tuning")
plt.xlabel('Months')
plt.ylabel('Price(U$ Dollar)')

plt.legend(labels=['Turbo', 'Davinci', 'Fine Tuned'])

plt.show()
```
