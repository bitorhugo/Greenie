```python
import matplotlib.pyplot as plt
```

```python
# price per 1K tokens
davinci_1k_price = 0.02
# fine-tuned price
ft_davinci_1k_price = 0.12
```

```python
# one question deals on average 1.5k-1.8k tokens (input/output combined)
# price per request = (totalTokens/1kToken) * model_price
price_per_request_davinci = (( 1800 / 1000 ) * davinci_1k_price)
ft_price_per_request_davinci = (( 1800 / 1000 ) * ft_davinci_1k_price)
print(f'Davinci per request: {price_per_request_davinci} $')
print(f'FineTuned per request: {ft_price_per_request_davinci} $')
```

```python
# let' say we have 20K customers a month ( 1/10 of OPorto Population )
# each deals 15 requests per month
monthly_requests = 300000
```

```python
# price of monthly requests
davinci_monthly = price_per_request_davinci * monthly_requests
ft_monthly = ft_price_per_request_davinci * monthly_requests
print(f'Davinci Monthly Expenses: {davinci_monthly}')
print(f'Turbo Monthly Expenses: {ft_monthly}')
```

```python
# calculate for 12 months
davinci_prices = []
ft_prices = []
for i in range (1, 13):
    davinci_prices.append(int(davinci_monthly * i))
    ft_prices.append(int(ft_monthly * i))
```

```python
plt.title("Fine Tuning Comparison")
plt.xlabel("Months")
plt.ylabel("Price(U$ Dollar)")
plt.xlim(1, 12)
plt.plot([i for i in range(1,13)],davinci_prices, label='davinci')
plt.plot([i for i in range(1,13)],ft_prices, label='fine-tuned-davinci')
plt.legend()
```
